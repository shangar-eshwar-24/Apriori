```python
tran = []
for i in range(len(data)):
  tran.append([str(data.values[i,j]) for j in range(20) if str(data.values[i,j]) != '0'])
```

```python
from apyori import apriori
```

```python
rules = apriori(transactions = tran, min_support=0.003, min_confidence=0.2, min_lift=3, min_length=2, max_length=2)
```

```python
results = list(rules)
```

```python
def inspect(results):
    lhs         = [result[2][0][0] for result in results]
    rhs         = [result[2][0][1] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))

results_df = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
print(results_df)
```
